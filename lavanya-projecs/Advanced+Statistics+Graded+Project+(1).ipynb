3.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mean = 5  # Mean breaking strength (kg per sq cm)
std_dev = 1.5  # Standard deviation (kg per sq cm)
x_value = 3.17  # Value to find the proportion for

# Calculate the z-score
z_score = (x_value - mean) / std_dev

# Calculate the cumulative probability
proportion_less_than_x = norm.cdf(z_score)

# Create a visual representation
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y = norm.pdf(x, mean, std_dev)
plt.plot(x, y, color='blue', label='Normal Distribution')
plt.fill_between(x, y, where=(x < x_value), alpha=0.3, color='orange', label=f'Area < {x_value} kg per sq cm')
plt.xlabel('Breaking Strength (kg per sq cm)')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution')
plt.text(4.8, 0.15, f'Proportion: {proportion_less_than_x:.4f}', fontsize=12, color='red')
plt.show()

```

3.2 What proportion of the gunny bags have a breaking strength at least 3.6 kg per sq cm.?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mean = 5  # Mean breaking strength (kg per sq cm)
std_dev = 1.5  # Standard deviation (kg per sq cm)
x_value = 3.6  # Value to find the proportion for

# Calculate the z-score
z_score = (x_value - mean) / std_dev

# Calculate the complementary cumulative probability
proportion_at_least_x = 1 - norm.cdf(z_score)

# Create a visual representation
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y = norm.pdf(x, mean, std_dev)
plt.plot(x, y, color='blue', label='Normal Distribution')
plt.fill_between(x, y, where=(x >= x_value), alpha=0.3, color='orange', label=f'Area >= {x_value} kg per sq cm')
plt.xlabel('Breaking Strength (kg per sq cm)')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution')
plt.text(4.8, 0.15, f'Proportion: {proportion_at_least_x:.4f}', fontsize=12, color='red')
plt.show()

```

3.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mean = 5  # Mean breaking strength (kg per sq cm)
std_dev = 1.5  # Standard deviation (kg per sq cm)
lower_limit = 5  # Lower limit of breaking strength (kg per sq cm)
upper_limit = 5.5  # Upper limit of breaking strength (kg per sq cm)

# Calculate z-scores
z_lower = (lower_limit - mean) / std_dev
z_upper = (upper_limit - mean) / std_dev

# Calculate cumulative probabilities
prob_lower = norm.cdf(z_lower)
prob_upper = norm.cdf(z_upper)

# Calculate the proportion
proportion_between_limits = prob_upper - prob_lower

# Create a visual representation
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y = norm.pdf(x, mean, std_dev)
plt.plot(x, y, color='blue', label='Normal Distribution')
plt.fill_between(x, y, where=((x >= lower_limit) & (x <= upper_limit)), alpha=0.3, color='orange', label=f'Area between {lower_limit} and {upper_limit} kg per sq cm')
plt.xlabel('Breaking Strength (kg per sq cm)')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution')
plt.text(4.8, 0.15, f'Proportion: {proportion_between_limits:.4f}', fontsize=12, color='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mean = 5  # Mean breaking strength (kg per sq cm)
std_dev = 1.5  # Standard deviation (kg per sq cm)
lower_limit = 3  # Lower limit of breaking strength (kg per sq cm)
upper_limit = 7.5  # Upper limit of breaking strength (kg per sq cm)

# Calculate z-scores
z_lower = (lower_limit - mean) / std_dev
z_upper = (upper_limit - mean) / std_dev

# Calculate cumulative probabilities
prob_lower = norm.cdf(z_lower)
prob_upper = norm.cdf(z_upper)

# Calculate the proportion
proportion_not_between_limits = 1 - (prob_upper - prob_lower)

# Create a visual representation
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y = norm.pdf(x, mean, std_dev)
plt.plot(x, y, color='blue', label='Normal Distribution')
plt.fill_between(x, y, where=((x < lower_limit) | (x > upper_limit)), alpha=0.3, color='orange', label=f'Area not between {lower_limit} and {upper_limit} kg per sq cm')
plt.xlabel('Breaking Strength (kg per sq cm)')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution')
plt.text(4.8, 0.15, f'Proportion: {proportion_not_between_limits:.4f}', fontsize=12, color='red')
plt.show()

```

Problem 4. Grades of the final examination in a training course are found to be normally distributed, with a mean of 77 and a standard deviation of 8.5. Based on the given information answer the questions below.

 

4.1 What is the probability that a randomly chosen student gets a grade below 85 on this exam?

4.2 What is the probability that a randomly selected student scores between 65 and 87?

4.3 What should be the passing cut-off so that 75% of the students clear the exam?

```python
# Given parameters
mean = 77  # Mean grade
std_dev = 8.5  # Standard deviation
grade_value = 85  # Grade value to find the probability for

# Calculate the z-score
z_score = (grade_value - mean) / std_dev

# Calculate the cumulative probability
probability_below_grade = norm.cdf(z_score)

# Print the probability
print(f"Probability that a randomly chosen student gets a grade below {grade_value}: {probability_below_grade:.4f}")

# Create a visual representation
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y = norm.pdf(x, mean, std_dev)
plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, y, where=(x < grade_value), alpha=0.3, color='orange', label=f'Area below {grade_value}')
plt.xlabel('Grades')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution')
plt.legend()
plt.show()
```

4.2 What is the probability that a randomly selected student scores between 65 and 87?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mean = 77  # Mean grade
std_dev = 8.5  # Standard deviation
lower_limit = 65  # Lower limit of grade
upper_limit = 87  # Upper limit of grade

# Calculate z-scores
z_lower = (lower_limit - mean) / std_dev
z_upper = (upper_limit - mean) / std_dev

# Calculate cumulative probabilities
prob_lower = norm.cdf(z_lower)
prob_upper = norm.cdf(z_upper)

# Calculate the probability of being between the limits
probability_between_limits = prob_upper - prob_lower

# Create a visual representation
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y = norm.pdf(x, mean, std_dev)
plt.plot(x, y, color='blue')
plt.fill_between(x, y, where=((x >= lower_limit) & (x <= upper_limit)), alpha=0.3, color='orange')
plt.xlabel('Grades')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution')
plt.text(72, 0.03, f'Probability: {probability_between_limits:.4f}', fontsize=12, color='red')
plt.show()
```

```python
# Given parameters
mean = 77  # Mean grade
std_dev = 8.5  # Standard deviation
percentile = 75  # Desired percentile

# Calculate the z-score corresponding to the desired percentile
z_percentile = norm.ppf(percentile / 100)

# Calculate the passing cut-off
passing_cutoff = mean + z_percentile * std_dev

# Print the passing cut-off
print(f"The passing cut-off for {percentile}% of students: {passing_cutoff:.2f}")

# Create a visual representation
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y = norm.pdf(x, mean, std_dev)
plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, y, where=(x >= passing_cutoff), alpha=0.3, color='orange', label=f'Passing cut-off')
plt.axvline(x=passing_cutoff, color='red', linestyle='--', label='Passing Cut-off')
plt.xlabel('Grades')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution')
plt.legend()
plt.show()
```

Zingaro stone printing is a company that specializes in printing images or patterns on polished or unpolished stones. However, for the optimum level of printing of the image the stone surface has to have a Brinell's hardness index of at least 150. Recently, Zingaro has received a batch of polished and unpolished stones from its clients. Use the data provided to answer the following (assuming a 5% significance level);

 

5.1 Earlier experience of Zingaro with this particular client is favorable as the stone surface was found to be of adequate hardness. However, Zingaro has reason to believe now that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

5.2 Is the mean hardness of the polished and unpolished stones the same?

 

```python
import pandas as pd
zing=pd.read_csv(r"C:\Users\lavan\Downloads\Zingaro_Company.csv")
```

```python
zing.head()
```

```python
zing.shape
```

```python

# Parameters
alpha = 0.05  # Significance level
mu_0 = 150    # Hypothesized minimum hardness for printing suitability

# Extract the 'Unpolished' column from the DataFrame
unpolished_hardness = zing['Unpolished ']

# Calculate sample size, mean, and standard deviation for unpolished stones
n = len(unpolished_hardness)
x_bar = unpolished_hardness.mean()
s = unpolished_hardness.std()

# Calculate t-statistic
t_statistic = (x_bar - mu_0) / (s / (n**0.5))


# Calculate degrees of freedom
df = len(zing['Unpolished ']) - 1

print("Degrees of Freedom:", df)
# Calculate critical t-value
critical_t_value = stats.t.ppf(1 - alpha, df)

# Perform the one-sample t-test
p_value = stats.t.cdf(t_statistic, df)

# Compare the t-statistic with the critical t-value and print results
if t_statistic < critical_t_value:
    print("Reject the null hypothesis: Unpolished stones are not suitable for printing.")
else:
    print("Fail to reject the null hypothesis: Unpolished stones may be suitable for printing.")

print("t-statistic:", t_statistic)
print("Critical t-value:", critical_t_value)
print("p-value:", p_value)
```

5.2 Is the mean hardness of the polished and unpolished stones the same?

```python
from scipy.stats import ttest_ind

# Extract hardness values for polished and unpolished stones
polished_hardness = zing['Treated and Polished']
unpolished_hardness = zing['Unpolished ']

# Perform independent t-test
t_statistic, p_value = ttest_ind(polished_hardness, unpolished_hardness)

alpha = 0.05  # Significance level

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in mean hardness between polished and unpolished stones.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in mean hardness between polished and unpolished stones.")
```

7.4. Now test whether there is any difference among the methods on the hardness of dental implant, separately for the two types of alloys. What are your conclusions? If the null hypothesis is rejected, is it possible to identify which pairs of methods differ?

```python
from scipy.stats import f_oneway

alloy1_df = df[df['Alloy'] == 1]
alloy2_df = df[df['Alloy'] == 2]

methods_alloy1 = alloy1_df['Method']
methods_alloy2 = alloy2_df['Method']

response_alloy1 = alloy1_df['Response']
response_alloy2 = alloy2_df['Response']

# Perform one-way ANOVA for Alloy 1
anova_result_alloy1 = f_oneway(response_alloy1[methods_alloy1 == 1],
                                response_alloy1[methods_alloy1 == 2],
                                response_alloy1[methods_alloy1 == 3])

print("ANOVA Results for Alloy 1:")
print("Null Hypothesis (H0): There is no difference among the methods on the hardness of dental implants for Alloy 1.")
print("Alternative Hypothesis (HA): There is a significant difference among the methods on the hardness of dental implants for Alloy 1.")
print("F-statistic:", anova_result_alloy1.statistic)
print("p-value:", anova_result_alloy1.pvalue)

# Perform one-way ANOVA for Alloy 2
anova_result_alloy2 = f_oneway(response_alloy2[methods_alloy2 == 1],
                                response_alloy2[methods_alloy2 == 2],
                                response_alloy2[methods_alloy2 == 3])

print("\nANOVA Results for Alloy 2:")
print("Null Hypothesis (H0): There is no difference among the methods on the hardness of dental implants for Alloy 2.")
print("Alternative Hypothesis (HA): There is a significant difference among the methods on the hardness of dental implants for Alloy 2.")
print("F-statistic:", anova_result_alloy2.statistic)
print("p-value:", anova_result_alloy2.pvalue)
```
